```python
# bussiness churn model
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from plotnine import *

#models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.compose import ColumnTransformer # for concatenating continuous and categorical variables
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import CategoricalNB

#preprocessing and validation
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, roc_auc_score 
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay


#load data
churnDF = pd.read_csv('https://raw.githubusercontent.com/katherinehansen2/CPSC392Hansen/main/data/streaming.csv')

churnDF.dropna(inplace = True)
churnDF.reset_index(inplace = True, drop = True)

#predictors
predictors = ['gender','age','income','monthssubbed','plan','meanhourswatched','competitorsub','numprofiles','cancelled','downgraded','bundle','kids','longestsession','topgenre','secondgenre']

X = churnDF[predictors]
y = churnDF['churn']

#train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)

#preprocessing
continousVars = ['age','income','monthssubbed','meanhourswatched','longestsession']
categoricalVars= ['gender','kids','numprofiles' ,'bundle', 'downgraded', 'cancelled', 'secondgenre', 'topgenre', 'plan', 'competitorsub']

#create column transformer
preprocess = ColumnTransformer(transformers = [("num", StandardScaler(), continousVars),
                                               ("cat", OneHotEncoder(drop = 'first'), categoricalVars)])

#create logistic regression pipeline model 1
lr = Pipeline(steps = [("preprocess", preprocess), ("classifier", LogisticRegression())])

lr.fit(X_train, y_train)
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)

print(ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred))
print("Train Accuracy: ", accuracy_score(y_train, y_train_pred))
print("Test Accuracy: ", accuracy_score(y_test, y_test_pred))
print("Precision Score: ", precision_score(y_test, y_test_pred))
print("Recall Score: ", recall_score(y_test, y_test_pred))
print("AUC: ", roc_auc_score(y_test, y_test_pred))

churnDF.head()

```

```python


#model 2 - gradient boosting
gb = Pipeline(steps = [("preprocess", preprocess), ("classifier", GradientBoostingClassifier(random_state = 123))])
gb.fit(X_train, y_train)
y_train_pred_gb = gb.predict(X_train)
y_test_pred_gb = gb.predict(X_test)

print("Gradient Boosting results")
print(ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred_gb))
print(ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred_gb))
print("Train Accuracy: ", accuracy_score(y_train, y_train_pred_gb))
print("Test Accuracy: ", accuracy_score(y_test, y_test_pred_gb))
print("Precision Score: ", precision_score(y_test, y_test_pred_gb))
print("Recall Score: ", recall_score(y_test, y_test_pred_gb))
print("AUC: ", roc_auc_score(y_test, y_test_pred_gb))

#churn prediction by top genre

topgenres = X_test['topgenre'].value_counts().index[:5] # top 5 genres
X_test_top = X_test[X_test['topgenre'].isin(topgenres)]

y_test_top_pred_gb = gb.predict(X_test_top)

ggplot_data = pd.DataFrame({"topgenre": X_test_top['topgenre'], "churn_pred": y_test_top_pred_gb})

(ggplot(ggplot_data, aes(x = "topgenre", fill = "factor(churn_pred)")) + 
 geom_bar(position="dodge") + 
 labs(title = "Churn Prediction by Top Genre", x = "Top Genre", y = "Count") + 
 theme_bw() + 
 theme(axis_text_x=element_text(rotation=45, hjust=1)))

```

```python
#Mushroom data
mushroomDf = pd.read_csv("mushrooms.csv")
mushroomDf.dropna(inplace = True)
mushroomDf.reset_index(inplace = True, drop = True)

#features and target
X = mushroomDf.drop(columns =[ 'class'])
y = mushroomDf['class']

#train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)  

#preprocessing
encoder = OneHotEncoder(sparse_output = False)
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)

#naive bayes model
nb = CategoricalNB()
nb.fit(X_train_encoded, y_train)

y_train_pred_nb = nb.predict(X_train_encoded) # added new
y_test_pred_nb = nb.predict(X_test_encoded)



print("Naive Bayes results")
print(ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred_nb))
print(ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred_nb))
print("Train Accuracy: ", accuracy_score(y_train, y_train_pred_nb))
print("Test Accuracy: ", accuracy_score(y_test, y_test_pred_nb))
print("Precision Score: ", precision_score(y_test, y_test_pred_nb, pos_label = 'p'))
print("Recall Score: ", recall_score(y_test, y_test_pred_nb, pos_label = 'p'))
print("AUC: ", roc_auc_score(y_test, nb.predict_proba(X_test_encoded)[:,1]))


#knn model
knn = KNeighborsClassifier()
knn.fit(X_train_encoded, y_train)

y_train_pred_knn = knn.predict(X_train_encoded)
y_test_pred_knn = knn.predict(X_test_encoded)

print("KNN results")
print(ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred_knn))
print(ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred_knn))
print("Train Accuracy: ", accuracy_score(y_train, y_train_pred_knn))
print("Test Accuracy: ", accuracy_score(y_test, y_test_pred_knn))
print("Precision Score: ", precision_score(y_test, y_test_pred_knn, pos_label = 'p'))
print("Recall Score: ", recall_score(y_test, y_test_pred_knn, pos_label = 'p'))
print("AUC: ", roc_auc_score(y_test, knn.predict_proba(X_test_encoded)[:,1]))

#logistic regression model
lr = LogisticRegression()
lr.fit(X_train_encoded, y_train)

y_train_pred_lr = lr.predict(X_train_encoded)
y_test_pred_lr = lr.predict(X_test_encoded)

print("Logistic Regression results")
print(ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred_lr))
print(ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred_lr))
print("Train Accuracy: ", accuracy_score(y_train, y_train_pred_lr))
print("Test Accuracy: ", accuracy_score(y_test, y_test_pred_lr))
print("Precision Score: ", precision_score(y_test, y_test_pred_lr, pos_label = 'p'))
print("Recall Score: ", recall_score(y_test, y_test_pred_lr, pos_label = 'p'))
print("AUC: ", roc_auc_score(y_test, lr.predict_proba(X_test_encoded)[:,1]))

```

```python
#poisonous mushrooms detector

def mushroomDetector(features,model):
    #encode the features
    features_encoded = encoder.transform([features])

    prediction = model.predict(features_encoded)

    if prediction == 'p':
        return "This mushroom is poisonous!"
    else:
        return "This mushroom is edible"
    
print(mushroomDetector(X_test.iloc[1], nb))
print(mushroomDetector(X_test.iloc[1], knn))
print(mushroomDetector(X_test.iloc[1], lr))
```
